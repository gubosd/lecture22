### 선형회귀 적용
- X : Logged GDP per capita
- y : Ladder score

```python
from sklearn.linear_model import LinearRegression
```

```python
X = whr[['Logged GDP per capita']].values # 2차원 행렬로 만들어야 함
X.shape
```

```python
y = whr['Ladder score'].values
y
```

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
y_pred
```

```python
model.score(X, y)
```

```python
xs = np.array([6, 12])
ys = model.coef_[0]*xs + model.intercept_
plt.scatter(X[:,0], y)
plt.plot(xs, ys, 'r')
```

- 실제 GDP 값 적용

- 자유도와 행복의 관계

```python
X = whr[['Freedom to make life choices']].values # 2차원 행렬로 만들어야 함
X.shape
```

```python
y = whr['Ladder score'].values
```

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
y_pred
```

```python
model.score(X, y)
```

```python
xs = np.array([0.3, 1])
ys = model.coef_[0]*xs + model.intercept_
plt.scatter(X[:,0], y)
plt.plot(xs, ys, 'r')
```
